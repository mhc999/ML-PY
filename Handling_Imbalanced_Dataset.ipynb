The dataset is imbalanced i.e., the number of values for True might be way greater than the number of values available for False

```python
import numpy as np
import pandas as pd
```

```python
# load the dataset to pandas dataframe
# this data is used to predict whether the transaction is legit or not
credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
credit_card_data.head()
# 0 = legit
```

```python
credit_card_data.tail()
```

```python
credit_card_data['Class'].value_counts()
# This is highly imbalanced dataset
```

```python
# seperate the legit and fraud data
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

UNDERSAMPLING :
Take a random sample of data points similar to the other value

```python
legit_sample = legit.sample(n=492)
```

```python
print(legit_sample.shape)
```

```python
# add the legit_sample and fraud to a new dataframe where the data is now not imbalanced
new_df = pd.concat([legit_sample, fraud] , axis = 0)
# axis = 0 --> the two data sets will be added row wise one after the other
# axis = 0 --> the two data sets will be added column wise one after the other
```
