Pandas Library : used for data processing and anslysis


Pandas Dataframe : 2 dim tabular data structure with labelled rows and columns

```python
#import the pandas library
import numpy as np
import pandas as pd
```

```python
#CREATE A PANDAS DATAFRAME
#import the boston house price data
from sklearn.datasets import load_boston
```

```python
boston_dataset = load_boston()
type(boston_dataset)

# type: Bunch --> bunch of data 
```

```python
print(boston_dataset)

# pandas helps us to store all this data in a structured table for easy analysis
```

```python
#pandas dataframe
# pd.DataFrame = creates a pandas dataframe 
# boston_dataset.data : load all the data into the data frame
# columns = boston_dataset.feature_names : columns should be named after the feature_names from the boston_dataset
boston_df = pd.DataFrame(boston_dataset.data , columns = boston_dataset.feature_names)
```

```python
#get the sample ( first five rows of the dataframe)
boston_df.head()
```

```python
# shape of data frame
boston_df.shape
```

MANIPULATING A DATAFRAME

```python
#adding a column to the DataFrame
boston_df['Price'] = boston_dataset.target
#boston_df['name of the new column'] = boston_dataset.name of the column from the dataset

boston_df.head()
```

```python
#removing a row 
boston_df.drop(index = 0, axis = 0)
# for the first row index = 0 and for rows, mention axis = 0. For columns, axis = 1 
```

```python
# removing a column
boston_df.drop(columns='ZN', axis = 1)
```

```python
# locating a row using the index value
boston_df.iloc[2]
#boston_df.iloc[ index of the row ]
```

```python
#locating a particular column
print(boston_df.iloc[:,0]) #first column
print(boston_df.iloc[:,1]) #second column
print(boston_df.iloc[:,2]) #third column
print(boston_df.iloc[:,-1]) # last column 
```
